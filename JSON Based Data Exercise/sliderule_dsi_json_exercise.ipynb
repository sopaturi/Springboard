## JSON example, with string

+ demonstrates creation of normalized dataframes (tables) from nested json string
+ source: http://pandas.pydata.org/pandas-docs/stable/io.html#normalization

```python
# define json string
data = [{'state': 'Florida', 
         'shortname': 'FL',
         'info': {'governor': 'Rick Scott'},
         'counties': [{'name': 'Dade', 'population': 12345},
                      {'name': 'Broward', 'population': 40000},
                      {'name': 'Palm Beach', 'population': 60000}]},
        {'state':'Ohio',
         'shortname': 'OH',
         'info': {'governor': 'John Kasich'},
         'counties': [{'name': 'Summit', 'population': 1234},
                      {'name': 'Cuyahoga', 'population': 1337}]}]
```

```python
# use normalization to create tables from nested element
json_normalize(data, 'counties')
```

```python
# further populate tables created from nested element
json_normalize(data, 'counties', ['state', 'shortname', ['info', 'governor']])

```

```python
#Exercise 1
sample_json_df.groupby('countryname').size().sort_values(ascending=False).head(10) #groupby top countries
```

```python
#Exercise 2
sample=pd.DataFrame(sample_json_df['mjtheme_namecode'])
sample=sample.mjtheme_namecode[0:(len(sample[('mjtheme_namecode')]))]  #isolate one field

codes={}

for themes in sample:      #loop through project types for each project
    for theme in themes:    #look through each project type
        if (theme['code'] in codes) | (theme['name'] == ''):  #make a dict of project types
            continue
        else:   #add to dict
            codes[theme['code']]=theme['name']
            
            

theme_list=[]

for t in sample:    #loop through project types for each project
        for s in t:  #look through each project type
            if s['name']=='':
                s['name']=codes[s['code']]   #replace misssing values
                theme_list.append((s['name']))
            else:
                theme_list.append((s['name']))
                
            
                
            
sample1=pd.DataFrame(theme_list)  #make dataframe
sample1=sample1.rename(columns={0: 'Top 10 Major Project Themes'})
sample1=sample1.groupby('Top 10 Major Project Themes').size().sort_values(ascending=False).head(10)
sample1


```
